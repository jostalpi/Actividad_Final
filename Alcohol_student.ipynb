```python
df4=pd.concat([df1,df2], ignore_index=True)
```

```python
df4.columns
```

```python

# Contar valores no nulos por grupo
count_por_grupo = df4.groupby('age')['Walc'].count()
print(count_por_grupo)
```

```python
edad_counts = count_por_grupo 
plt.figure(figsize=(10, 6))
plt.rc('font', size=8)
edad_counts.plot(kind='bar', color='skyblue')
plt.title('Consumo de alcohol por edades')
plt.xlabel('Edad')
plt.ylabel('Consumo de alcohol fin de semana') 
plt.show()
```

```python
mapeo = {
    1: 'Muy mal',
    2: 'Mal',
    3: 'Regular',
    4: 'Bueno',
    5: 'Muy bueno'
}
df4['health1'] = df4['health'].map(mapeo)
estado_de_salud = df4.groupby('health1')['Walc'].count()
print(estado_de_salud)
```
